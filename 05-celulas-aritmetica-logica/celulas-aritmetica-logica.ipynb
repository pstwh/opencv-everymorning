```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('celulas.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.imshow(img1)
```

```python
img2 = cv2.imread('celulas2.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = img2[:375, :500]

plt.imshow(img2)
```

```python
imgSoma = img1 + img2

plt.imshow(imgSoma)
```

```python
imgAdicionar = cv2.add(img1, img2)

plt.imshow(imgAdicionar)
```

```python
imgAdicionarComPeso = cv2.addWeighted(img1, 0.15, img2, 0.9, 0)

plt.imshow(imgAdicionarComPeso)
```

```python
imgIcone = cv2.imread('icone.png', cv2.IMREAD_UNCHANGED)
imgIcone = cv2.cvtColor(imgIcone, cv2.COLOR_BGR2RGB)

plt.imshow(imgIcone)
```

```python
imgIconeGray = cv2.cvtColor(imgIcone, cv2.COLOR_RGB2GRAY)

plt.imshow(imgIconeGray, 'gray')
```

```python
ret, imgIconeMascara = cv2.threshold(imgIconeGray, 200, 255, cv2.THRESH_BINARY)
imgIconeMascara = cv2.medianBlur(imgIconeMascara, 3)

plt.imshow(imgIconeMascara, 'binary')
```

```python
imgIconeMascaraInvertida = cv2.bitwise_not(imgIconeMascara)

plt.imshow(imgIconeMascaraInvertida, 'binary')
```

```python
linhasImgIcone, colunasImgIcone, canaisImgIcone = imgIcone.shape

localImg1 = imgAdicionarComPeso[:linhasImgIcone, :colunasImgIcone]

imgFundo = cv2.bitwise_and(localImg1, localImg1, mask=imgIconeMascaraInvertida)

plt.imshow(imgFundo)
```

```python
imgTopo = cv2.bitwise_and(imgIcone, imgIcone, mask=imgIconeMascara)

plt.imshow(imgTopo)
```

```python
imgJuncao = cv2.add(imgFundo, imgTopo)

plt.imshow(imgJuncao)
```

```python
imgAdicionarComPeso[:linhasImgIcone, :colunasImgIcone] = imgJuncao

plt.imshow(imgAdicionarComPeso)
```
