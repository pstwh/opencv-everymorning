```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
imgOriginal = cv2.imread('CelB1.bmp')

plt.imshow(imgOriginal)
```

```python
img = cv2.medianBlur(imgOriginal, 3)

plt.imshow(img)
```

```python
cv2.line(img, (0,0), (500, 500), (255, 255, 255), 5)

plt.imshow(img)
```

```python
cv2.rectangle(img, (200,200), (350, 350), (255, 255, 255), 5)

plt.imshow(img)
```

```python
cv2.circle(img, (200, 200), 75, (255, 255, 255), 5)

plt.imshow(img)
```

```python
polygon = np.array([[200, 200], [150, 200], [150, 500]], np.int32)

cv2.polylines(img, [polygon], False, (255, 255, 255), 5)

plt.imshow(img)
```

```python
text = "Hello"
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(img, text, (210, 340), font, 1, (255, 255, 255), 2, cv2.LINE_AA)

plt.imshow(img)
```
