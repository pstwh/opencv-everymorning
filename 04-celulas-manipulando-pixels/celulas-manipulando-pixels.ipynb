```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
imgOriginal = cv2.imread('celulas.jpg')
imgOriginal = cv2.cvtColor(imgOriginal, cv2.COLOR_BGR2RGB)

plt.imshow(imgOriginal)
```

```python
imgGrayscale = cv2.cvtColor(imgOriginal, cv2.COLOR_BGR2GRAY)

plt.imshow(imgGrayscale, 'gray')
```

```python
ret, imgBinary = cv2.threshold(imgGrayscale, 110, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(imgBinary, 'binary_r')
```

```python
imgBinary[:200, :200] = 0

plt.imshow(imgBinary, 'binary_r')

print("Pixel na posição (200, 350) tem a coloração: {}".format(imgBinary[350, 200]))
print("O tamanho da imagem é: {}".format(imgBinary.shape))
print("A quantidade de pixels da imagem é: {}".format(imgBinary.size))
print("O tipo utilizado nos pixels da imagem é: {}".format(imgBinary.dtype))
```
