```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
imgOriginal = cv2.imread('page.jpg')
imgOriginal = cv2.cvtColor(imgOriginal, cv2.COLOR_BGR2RGB)

plt.imshow(imgOriginal)
```

```python
ret, imgThresh1 = cv2.threshold(imgOriginal, 10, 255, cv2.THRESH_BINARY)

plt.imshow(imgThresh1)
```

```python
ret, imgThresh2 = cv2.threshold(imgOriginal, 12, 255, cv2.THRESH_BINARY)

plt.imshow(imgThresh2)
```

```python
imgGrayscale = cv2.cvtColor(imgOriginal, cv2.COLOR_RGB2GRAY)

plt.imshow(imgGrayscale, 'gray')
```

```python
ret, imgThresh3 = cv2.threshold(imgGrayscale, 10, 255, cv2.THRESH_BINARY)

plt.imshow(imgThresh3, 'binary_r')
```

```python
ret, imgThresh4 = cv2.threshold(imgGrayscale, 12, 255, cv2.THRESH_BINARY)

plt.imshow(imgThresh4, 'binary_r')
```

```python
ret, imgThresh5 = cv2.threshold(imgGrayscale, 115, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(imgThresh5, 'binary_r')
```

```python
imgThresh6 = cv2.adaptiveThreshold(imgGrayscale, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)

plt.imshow(imgThresh6, 'binary_r')

"""
cv2.imshow('Threshold Adaptativo',imgThresh6)
cv2.waitKey(0)
cv2.destroyAllWindows()
"""
```
